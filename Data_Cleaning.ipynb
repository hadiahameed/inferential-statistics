```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pandas as pd
```

```python
og_data = pd.read_csv('data/Citywide_Payroll_Data__Fiscal_Year_.csv')
```

```python
display(og_data.head())
```

```python
data = og_data[(og_data['Fiscal Year'] == 2020) | (og_data['Fiscal Year'] == 2021)][['Fiscal Year','First Name', 'Last Name','Agency Name','Work Location Borough', 'Title Description', 'Base Salary']]
data.head()
```

```python
data = data.drop_duplicates(subset=["First Name", "Last Name", "Fiscal Year"])
```

```python
print(data['Agency Name'].value_counts().to_string())
```

```python
police_data = data[data['Agency Name'] == 'POLICE DEPARTMENT']
police_data.reset_index(drop=True, inplace=True)
ss_data = data[data['Agency Name'] == 'HRA/DEPT OF SOCIAL SERVICES']
ss_data.reset_index(drop=True, inplace=True)

police_data.head()
```

```python
police_data_2020 = police_data[police_data['Fiscal Year'] == 2020]
police_data_2021 = police_data[police_data['Fiscal Year'] == 2021]
police_merged = pd.merge(police_data_2020, police_data_2021[['Base Salary','First Name', 'Last Name']], how='inner', on=('First Name','Last Name'), suffixes=('_2020', '_2021'))
police_merged.drop(['Fiscal Year'], axis=1, inplace=True)
police_merged['Employee_uuid'] = abs(police_merged[['First Name', 'Last Name']].sum(axis=1).map(hash))
police_merged = police_merged[['Employee_uuid','Agency Name','Work Location Borough','Title Description','Base Salary_2020','Base Salary_2021']]
police_merged.head()
```

```python
ss_data_2020 = ss_data[ss_data['Fiscal Year'] == 2020]
ss_data_2021 = ss_data[ss_data['Fiscal Year'] == 2021]
ss_merged = pd.merge(ss_data_2020, ss_data_2021[['Base Salary','First Name', 'Last Name']], how='inner', on=('First Name','Last Name'), suffixes=('_2020', '_2021'))
ss_merged.drop(['Fiscal Year'], axis=1, inplace=True)
ss_merged['Employee_uuid'] = abs(ss_merged[['First Name', 'Last Name']].sum(axis=1).map(hash))
ss_merged = ss_merged[['Employee_uuid','Agency Name','Work Location Borough','Title Description','Base Salary_2020','Base Salary_2021']]
ss_merged.head()
```

```python
# create a excel writer object
with pd.ExcelWriter("data/clean_data.xlsx") as writer:
   
    # use to_excel function and specify the sheet_name and index 
    # to store the dataframe in specified sheet
    police_merged.to_excel(writer, sheet_name="police", index=False)
    ss_merged.to_excel(writer, sheet_name="social_services", index=False)
```
